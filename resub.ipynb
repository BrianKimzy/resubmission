```python
import pandas as pd
import numpy as np
```

```python
#loading csv files
data = pd.read_csv("Telcom_dataset.csv")
data.head(5)
```

```python
dataa = pd.read_csv("Telcom_dataset2.csv")
dataa.head(5)
```

```python
dataaa = pd.read_csv("Telcom_dataset3.csv")
dataaa.head(5)
```

```python
cells = pd.read_csv("cells_geo.csv")
cells.head(10)
```

```python
#merge the tables
df1= pd.concat([data,dataa,dataaa,cells]).reset_index(drop=True)
df1.head()
```

```python
#Dropping duplicates
df1.drop_duplicates()
df1
```

```python
df3= df1.drop(['COUNTRY_A','COUNTRY_B','CELL_ON_SITE'], axis=1)
df3.head()
```

```python
df2= df3.groupby(['CELL_ID'], as_index=False)['PRODUCT','VALUE','VILLES'].agg(lambda x: list(x))
                  
         
df2
```

```python
df4= pd.value_counts(df3['VILLES'].values).iloc[:3]
df4
```
